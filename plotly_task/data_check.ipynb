```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('games.csv')
df.head(10)
```

```python
print("========describe========\n")
print(df.describe())
print("========dtypes========\n")
print(df.dtypes)
print("========info========\n")
print(df.info())
```

```python
print("======== Missing values========\n")
missing_values = df.isnull().sum()
missing_percent = (missing_values / len(df)) * 100
missing_data = pd.DataFrame({
    'q-ty of missing values': missing_values,
    'missing value rate(%)': missing_percent.round(2)
})
print(missing_data)
```

```python
plt.figure(figsize=(10, 5))
df['Year_of_Release'].value_counts().sort_index().plot(kind='bar')
plt.title('q-ty of games by year')
plt.xlabel('released year')
plt.ylabel('quantity of games')
plt.tight_layout()
plt.show()
```

```python
# Distribution by Genre
plt.figure(figsize=(10, 5))
df['Genre'].value_counts().plot(kind='bar')
plt.title('q-ty of games by genre')
plt.xlabel('genre')
plt.ylabel('games')
plt.tight_layout()
plt.show()
```
